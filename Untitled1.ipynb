```python
driver = gs.UndetectedChrome()
# Step 1: Navigate to the match page
url = "https://www.vlr.gg/459081/shopify-rebellion-black-vs-winthrop-university-challengers-league-2025-north-america-ace-stage-1/?game=206405&tab=overview"
driver.get(url)
time.sleep(3)  # Allow time for page to load

# Step 2: Click on the "Overview" tab
driver.find_element(By.LINK_TEXT, "Overview").click()
time.sleep(2)

# Step 3: Click on the first map
first_map = driver.find_element(By.XPATH, "//div[contains(@class, 'vm-stats-gamesnav-item') and contains(@class, 'js-map-switch')]")
first_map.click()
time.sleep(2)

# Step 4: Extract page content
soup = BeautifulSoup(driver.page_source, "html.parser")

# Step 5: Extract team names
teams = soup.find_all("div", class_="team-name")
team1 = teams[0].text.strip() if len(teams) > 0 else "N/A"
team2 = teams[1].text.strip() if len(teams) > 1 else "N/A"

# Step 6: Identify Attack or Defense Start
side_indicators = soup.find_all("div", class_="vlr-rounds-row")
team1_side, team2_side = "Unknown", "Unknown"

if len(side_indicators) >= 2:
    if "attack" in side_indicators[0].get("class", ""):
        team1_side, team2_side = "Attack", "Defense"
    else:
        team1_side, team2_side = "Defense", "Attack"

print(f"{team1} started on: {team1_side}")
print(f"{team2} started on: {team2_side}")

# Step 7: Extract round progression from "vlr-rounds"
rounds_section = soup.find("div", class_="vlr-rounds")

# Step 8: Extract round details with actual team names
round_progression = []
if rounds_section:
    round_rows = rounds_section.find_all("div", class_="vlr-rounds-row")  # Each row = Team's rounds

    for team_index, row in enumerate(round_rows):
        rounds = row.find_all("div", class_="vlr-rounds-row-col")
        round_data = [round.get("title") for round in rounds]  # Extract round info
        real_team_name = team1 if team_index == 0 else team2  # Assign correct team names
        round_progression.append({"Team": real_team_name, "Rounds": round_data})

# Step 9: Convert to DataFrame and Save
df_rounds = pd.DataFrame(round_progression)

# ✅ Assign Attack/Defense Info Correctly
df_rounds["Starting Side"] = df_rounds["Team"].apply(lambda x: team1_side if x == team1 else team2_side)

df_rounds.to_csv("vlr_rounds_progression.csv", index=False)

# Print extracted data
print(df_rounds)
```

```python
import time
from bs4 import BeautifulSoup
import pandas as pd
from selenium.webdriver.common.by import By
import google_colab_selenium as gs
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException

# Start Selenium
driver = gs.UndetectedChrome()
url = "https://vlr.gg/"
driver.get(url)

# Wait for the page to load
time.sleep(5)

# Click on the "Matches" tab
driver.find_element(By.LINK_TEXT, "Matches").click()
time.sleep(3)

# Click on the "Results" tab
driver.find_element(By.LINK_TEXT, "Results").click()
time.sleep(3)

# Get total number of pages
pagination = driver.find_elements(By.CSS_SELECTOR, "a[href*='?page=']")
total_pages = int(pagination[-1].text.strip()) if pagination else 1
print(f"Total Pages Found: {total_pages}")

# Initialize list to store all match data
all_matches_data = []

### **LOOP THROUGH MATCH PAGES**
current_page = 1
while current_page <= total_pages:
    print(f"\nScraping Page {current_page}/{total_pages}...")

    # Get updated match links each time before looping
    match_links = driver.find_elements(By.CSS_SELECTOR, "a.wf-module-item.match-item")

    # Loop through each match on the page
    for i in range(len(match_links)):
        try:
            match_links = driver.find_elements(By.CSS_SELECTOR, "a.wf-module-item.match-item")
            match = match_links[i]  # Get match at index i

            driver.execute_script("arguments[0].scrollIntoView();", match)  # Scroll to element if needed
            match.click()  # Click to open match page
            time.sleep(3)  # Wait for page to load

            # Extract match details
            match_soup = BeautifulSoup(driver.page_source, "html.parser")

            # Extract Team Names
            teams = match_soup.find_all("div", class_="team-name")
            team1_name = teams[0].text.strip() if len(teams) > 0 else "N/A"
            team2_name = teams[1].text.strip() if len(teams) > 1 else "N/A"

            # Extract Match Date
            date_element = match_soup.find("div", class_="match-header-date")
            match_date = date_element.text.strip().replace("\n", "").replace("\t", " ") if date_element else "N/A"
            match_date = " ".join(match_date.split())  # Remove duplicate spaces

            # Find all map sections
            map_sections = match_soup.find_all("div", class_="vm-stats-game")

            # Initialize list to store match data
            new_rows = []

            # Loop through each map section
            for map_section in map_sections:
                map_name = map_section.find("div", class_="map").text.strip() if map_section.find("div", class_="map") else "N/A"
                map_name = " ".join(map_name.split())  # Removes extra spaces, \n, \t

                # Extract scores
                scores = map_section.find_all("div", class_="score")
                if len(scores) == 2:
                    team1_map_score = int(scores[0].text.strip())
                    team2_map_score = int(scores[1].text.strip())
                    team1_win = 1 if team1_map_score > team2_map_score else 0
                else:
                    team1_map_score = "N/A"
                    team2_map_score = "N/A"
                    team1_win = "N/A"

                # Append match data
                new_rows.append({
                    "team1": team1_name,
                    "team2": team2_name,
                    "date": match_date,
                    "map_name": map_name,
                    "team1_map_score": team1_map_score,
                    "team2_map_score": team2_map_score,
                    "team1_win": team1_win
                })

            # Convert to DataFrame
            df = pd.DataFrame(new_rows)

            # Remove unplayed maps (where map_name or team1_win is "N/A")
            try:
              df = df[(df["map_name"] != "N/A") & (df["team1_win"] != "N/A")]
            except KeyError:
              print(f"⚠ Warning: Skipping match due to missing 'map_name' for {team1_name} vs {team2_name}")
              continue  # Skip this match

            ### **ECONOMY DATA EXTRACTION**
            try:
                # Click on the "Economy" tab
                econ_tab = driver.find_element(By.LINK_TEXT, "Economy")
                econ_tab.click()
                time.sleep(3)  # Allow the page to load

                # Get updated page source
                match_soup = BeautifulSoup(driver.page_source, "html.parser")

                # Locate all economy sections (ALL maps played)
                active_map_sections = match_soup.find_all("div", class_="vm-stats-game")  # No [:3] limit

                # Function to extract attempts & wins from "X (Y)" format
                def extract_attempts_wins(text):
                    text = text.strip().replace("\n", "").replace("\t", " ")  # Clean text
                    text = " ".join(text.split())  # Remove extra spaces
                    if "(" in text:
                        attempts, wins = text.split("(")[0].strip(), text.split("(")[1].replace(")", "").strip()
                    else:
                        attempts, wins = text, "0"  # If no parentheses, assume 0 wins
                    return int(attempts), int(wins)

                # Initialize list to store economy data
                all_econ_data = []

                # Loop through ALL map sections
                for idx, active_map_section in enumerate(active_map_sections):
                    econ_table = active_map_section.find("table", class_="wf-table-inset mod-econ")

                    if econ_table:
                        rows = econ_table.find_all("tr")

                        if len(rows) > 1:
                            first_team_row = rows[1]  # Index 1, since index 0 is the header
                            first_team_cols = [col.text.strip() for col in first_team_row.find_all("td")]

                            if len(first_team_cols) >= 6:
                                team_name = first_team_cols[0]
                                pistol_won = int(first_team_cols[1])

                                eco_attempted, eco_won = extract_attempts_wins(first_team_cols[2])
                                semi_eco_attempted, semi_eco_won = extract_attempts_wins(first_team_cols[3])
                                semi_buy_attempted, semi_buy_won = extract_attempts_wins(first_team_cols[4])
                                full_buy_attempted, full_buy_won = extract_attempts_wins(first_team_cols[5])

                                all_econ_data.append([
                                    idx + 1, team_name, pistol_won,
                                    eco_attempted, eco_won,
                                    semi_eco_attempted, semi_eco_won,
                                    semi_buy_attempted, semi_buy_won,
                                    full_buy_attempted, full_buy_won
                                ])

                df_econ = pd.DataFrame(all_econ_data)

                # Merge Economy Data with Match Data by Row Order
                df_final = pd.concat([df.reset_index(drop=True), df_econ.reset_index(drop=True)], axis=1)

            except:
                df_final = df  # If no economy data, just save match results

            all_matches_data.append(df_final)

            # **Auto-Save after every match**
            pd.concat(all_matches_data, ignore_index=True).to_csv("all_vlr_matches.csv", index=False)

            print(f"✔ Match {i+1} on Page {current_page} Saved ✅")
            driver.back()
            time.sleep(3)

        except StaleElementReferenceException:
            continue
        except NoSuchElementException:
            continue

    # Click Next Page
    try:
        next_button = driver.find_element(By.LINK_TEXT, str(current_page + 1))
        next_button.click()
        current_page += 1
        time.sleep(3)
    except:
        print("No more pages found. Exiting loop.")
        break

df_all_matches = pd.concat(all_matches_data, ignore_index=True)
df_all_matches.to_csv("all_vlr_matches.csv", index=False)
print("\n✅ All Matches Data Saved: all_vlr_matches.csv ✅")
```

```python
import re
import time
from bs4 import BeautifulSoup
import pandas as pd
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import google_colab_selenium as gs
from selenium.common.exceptions import TimeoutException, StaleElementReferenceException, NoSuchElementException

# Start Selenium
driver = gs.UndetectedChrome()
url = "https://vlr.gg/matches/results/?page=31"  # Start from page 18
driver.get(url)

# Get total number of pages
try:
    pagination = driver.find_elements(By.CSS_SELECTOR, "a[href*='?page=']")
    total_pages = int(pagination[-1].text.strip()) if pagination else 1
    print(f"Total Pages Found: {total_pages}")
except Exception as e:
    print(f"⚠ Error retrieving pagination: {e}")
    total_pages = 1

# **Correctly extract the current page number using regex**
try:
    match = re.search(r"page=(\d+)", driver.current_url)
    if match:
        current_page = int(match.group(1))  # Extract integer page number
    else:
        current_page = 1  # Default to page 1 if no match found
except:
    current_page = 1  # Default in case of any error

print(f"Starting scrape from page {current_page}/{total_pages}...")

# Initialize list to store match data
all_matches_data = []

### **LOOP THROUGH MATCH PAGES**
while current_page <= total_pages:
    print(f"\nScraping Page {current_page}/{total_pages}...")

    try:
        match_links = driver.find_elements(By.CSS_SELECTOR, "a.wf-module-item.match-item")
    except Exception as e:
        print(f"⚠ Error retrieving match links: {e}")
        continue

    for i in range(len(match_links)):
        try:
            match_links = driver.find_elements(By.CSS_SELECTOR, "a.wf-module-item.match-item")
            match = match_links[i]
            driver.execute_script("arguments[0].scrollIntoView();", match)
            match.click()
            WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.CLASS_NAME, "match-header-date")))

            match_soup = BeautifulSoup(driver.page_source, "html.parser")

            # Extract Team Names & Elo Ratings
            try:
                teams = match_soup.find_all("div", class_="team-name")
                elos = match_soup.find_all("div", class_="match-header-link-name-elo")

                team1_name = teams[0].text.strip() if len(teams) > 0 else "N/A"
                team2_name = teams[1].text.strip() if len(teams) > 1 else "N/A"

                team1_elo = elos[0].text.strip().replace("[", "").replace("]", "") if len(elos) > 0 else "N/A"
                team2_elo = elos[1].text.strip().replace("[", "").replace("]", "") if len(elos) > 1 else "N/A"
                forfeit_note = match_soup.find("div", class_="match-header-note")
                if forfeit_note and "forfeit" in forfeit_note.text.lower():
                    print(f"⚠ Skipping forfeited match: {team1_name} vs {team2_name}")
                    driver.back()
                    time.sleep(2)
            except Exception as e:
                print(f"⚠ Error extracting team names or elos: {e}")
                continue

            # Extract Match Date
            try:
                date_element = match_soup.find("div", class_="match-header-date")
                match_date = date_element.text.strip().replace("\n", "").replace("\t", " ") if date_element else "N/A"
                match_date = " ".join(match_date.split())
            except Exception as e:
                print(f"⚠ Error extracting match date: {e}")
                match_date = "N/A"

            # Find all maps played
            map_sections = match_soup.find_all("div", class_="vm-stats-game")

            new_rows = []
            for map_section in map_sections:
                try:
                    map_name = map_section.find("div", class_="map").text.strip() if map_section.find("div", class_="map") else "N/A"

                    scores = map_section.find_all("div", class_="score")
                    team1_map_score = int(scores[0].text.strip()) if len(scores) == 2 else "N/A"
                    team2_map_score = int(scores[1].text.strip()) if len(scores) == 2 else "N/A"
                    team1_win = 1 if team1_map_score != "N/A" and team1_map_score > team2_map_score else 0

                    new_rows.append({
                        "team1": team1_name, "team1_elo": team1_elo,
                        "team2": team2_name, "team2_elo": team2_elo,
                        "date": match_date, "map_name": map_name,
                        "team1_map_score": team1_map_score, "team2_map_score": team2_map_score,
                        "team1_win": team1_win
                    })
                except Exception as e:
                    print(f"⚠ Error extracting map data: {e}")
                    continue

            df = pd.DataFrame(new_rows)

            # Remove unplayed maps
            try:
                df = df[(df["map_name"] != "N/A") & (df["team1_win"] != "N/A")]
            except KeyError:
                print(f"⚠ Skipping match due to missing columns: {team1_name} vs {team2_name}")
                continue
             # **ECONOMY DATA EXTRACTION**
            try:
                econ_tab = driver.find_element(By.LINK_TEXT, "Economy")
                econ_tab.click()
                WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.CLASS_NAME, "mod-econ")))

                match_soup = BeautifulSoup(driver.page_source, "html.parser")
                active_map_sections = match_soup.find_all("div", class_="vm-stats-game")

                def extract_attempts_wins(text):
                    text = text.strip().replace("\n", "").replace("\t", " ")
                    text = " ".join(text.split())
                    if "(" in text:
                        attempts, wins = text.split("(")[0].strip(), text.split("(")[1].replace(")", "").strip()
                    else:
                        attempts, wins = text, "0"
                    return int(attempts), int(wins)

                all_econ_data = []
                for idx, active_map_section in enumerate(active_map_sections):
                    econ_table = active_map_section.find("table", class_="wf-table-inset mod-econ")
                    if econ_table:
                        rows = econ_table.find_all("tr")
                        if len(rows) > 1:
                            first_team_row = rows[1]
                            first_team_cols = [col.text.strip() for col in first_team_row.find_all("td")]

                            if len(first_team_cols) >= 6:
                                all_econ_data.append([
                                    idx + 1, first_team_cols[0], int(first_team_cols[1]),
                                    *extract_attempts_wins(first_team_cols[2]),
                                    *extract_attempts_wins(first_team_cols[3]),
                                    *extract_attempts_wins(first_team_cols[4]),
                                    *extract_attempts_wins(first_team_cols[5])
                                ])

                df_econ = pd.DataFrame(all_econ_data)
                df_final = pd.concat([df.reset_index(drop=True), df_econ.reset_index(drop=True)], axis=1)
            except Exception as e:
                print(f"⚠ Error extracting economy data: {e}")
                df_final = df

            all_matches_data.append(df_final)
            pd.concat(all_matches_data, ignore_index=True).to_csv("all_vlr_matches.csv", index=False)


            print(f"✔ Match {i+1} on Page {current_page} Saved ✅")
            driver.back()
            WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.CSS_SELECTOR, "a.wf-module-item.match-item")))

        except Exception as e:
            print(f"⚠ Skipping match due to error: {e}")
            driver.back()
            continue

    # **Click Next Page Button Correctly**
    try:
        next_button = WebDriverWait(driver, 5).until(
            EC.element_to_be_clickable((By.LINK_TEXT, str(current_page + 1)))
        )
        next_button.click()
        current_page += 1
        WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.CSS_SELECTOR, "a.wf-module-item.match-item")))

    except TimeoutException:
        print("⚠ No more pages found. Exiting loop.")
        break
    except StaleElementReferenceException:
        print("⚠ Stale element error. Refreshing page and retrying.")
        driver.refresh()
        time.sleep(3)
        continue
    except Exception as e:
        print(f"⚠ Error clicking next page: {e}")
        break

# Save final dataset
df_all_matches = pd.concat(all_matches_data, ignore_index=True)
df_all_matches.to_csv("all_vlr_matches.csv", index=False)
print("\n✅ All Matches Data Saved: all_vlr_matches.csv ✅")
```

```python
import time
from bs4 import BeautifulSoup
import pandas as pd
from selenium.webdriver.common.by import By
import google_colab_selenium as gs
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException

# Start Selenium
driver = gs.UndetectedChrome()
url = "https://vlr.gg/"
driver.get(url)

# Wait for the page to load
time.sleep(5)

# Click on the "Matches" tab
driver.find_element(By.LINK_TEXT, "Matches").click()
time.sleep(3)

# Click on the "Results" tab
driver.find_element(By.LINK_TEXT, "Results").click()
time.sleep(3)

# Get total number of pages
pagination = driver.find_elements(By.CSS_SELECTOR, "a[href*='?page=']")
total_pages = int(pagination[-1].text.strip()) if pagination else 1
print(f"Total Pages Found: {total_pages}")

# Initialize list to store all match data
all_matches_data = []

### **LOOP THROUGH MATCH PAGES**
current_page = 1
while current_page <= total_pages:
    print(f"\nScraping Page {current_page}/{total_pages}...")

    # Get updated match links each time before looping
    match_links = driver.find_elements(By.CSS_SELECTOR, "a.wf-module-item.match-item")

    # Loop through each match on the page
    for i in range(len(match_links)):
        try:
            match_links = driver.find_elements(By.CSS_SELECTOR, "a.wf-module-item.match-item")
            match = match_links[i]  # Get match at index i

            driver.execute_script("arguments[0].scrollIntoView();", match)  # Scroll to element if needed
            match.click()  # Click to open match page
            time.sleep(3)  # Wait for page to load

            # Extract match details
            match_soup = BeautifulSoup(driver.page_source, "html.parser")

            # Extract Team Names
            teams = match_soup.find_all("div", class_="team-name")
            team1_name = teams[0].text.strip() if len(teams) > 0 else "N/A"
            team2_name = teams[1].text.strip() if len(teams) > 1 else "N/A"

            # Extract Match Date
            date_element = match_soup.find("div", class_="match-header-date")
            match_date = date_element.text.strip().replace("\n", "").replace("\t", " ") if date_element else "N/A"
            match_date = " ".join(match_date.split())  # Remove duplicate spaces

            # Find all map sections
            map_sections = match_soup.find_all("div", class_="vm-stats-game")

            # Initialize list to store match data
            new_rows = []
            total_team1_wins = 0
            total_team2_wins = 0

            # Loop through each map section
            for map_section in map_sections:
                map_name = map_section.find("div", class_="map").text.strip() if map_section.find("div", class_="map") else "N/A"
                map_name = " ".join(map_name.split())  # Removes extra spaces, \n, \t

                # Extract scores and determine winner
                scores = map_section.find_all("div", class_="score")
                if len(scores) == 2:
                    team1_map_score = int(scores[0].text.strip())
                    team2_map_score = int(scores[1].text.strip())
                    team1_win = 1 if team1_map_score > team2_map_score else 0

                    # Track total wins for match score calculation
                    if team1_win:
                        total_team1_wins += 1
                    else:
                        total_team2_wins += 1
                else:
                    team1_win = "N/A"

                # Append match data
                new_rows.append({
                    "team1": team1_name,
                    "team2": team2_name,
                    "date": match_date,
                    "map_name": map_name,
                    "team1_win": team1_win
                })

            # Convert to DataFrame
            df = pd.DataFrame(new_rows)

            # Remove unplayed maps (where map_name or team1_win is "N/A")
            df = df[(df["map_name"] != "N/A") & (df["team1_win"] != "N/A")]

            # **Calculate match score**
            final_score = f"{total_team1_wins}-{total_team2_wins}"
            match_winner = team1_name if total_team1_wins > total_team2_wins else team2_name

            df["final_score"] = final_score
            df["match_winner"] = match_winner

            ### **ECONOMY DATA EXTRACTION**
            try:
                # Click on the "Economy" tab
                econ_tab = driver.find_element(By.LINK_TEXT, "Economy")
                econ_tab.click()
                time.sleep(3)  # Allow the page to load

                # Get updated page source
                match_soup = BeautifulSoup(driver.page_source, "html.parser")

                # Locate all economy sections (ALL maps played)
                active_map_sections = match_soup.find_all("div", class_="vm-stats-game")  # No [:3] limit

                # Function to extract attempts & wins from "X (Y)" format
                def extract_attempts_wins(text):
                    text = text.strip().replace("\n", "").replace("\t", " ")  # Clean text
                    text = " ".join(text.split())  # Remove extra spaces
                    if "(" in text:
                        attempts, wins = text.split("(")[0].strip(), text.split("(")[1].replace(")", "").strip()
                    else:
                        attempts, wins = text, "0"  # If no parentheses, assume 0 wins
                    return int(attempts), int(wins)

                # Initialize list to store economy data
                all_econ_data = []

                # Loop through ALL map sections
                for idx, active_map_section in enumerate(active_map_sections):
                    econ_table = active_map_section.find("table", class_="wf-table-inset mod-econ")

                    if econ_table:
                        rows = econ_table.find_all("tr")

                        if len(rows) > 1:
                            first_team_row = rows[1]  # Index 1, since index 0 is the header
                            first_team_cols = [col.text.strip() for col in first_team_row.find_all("td")]

                            if len(first_team_cols) >= 6:
                                team_name = first_team_cols[0]
                                pistol_won = int(first_team_cols[1])

                                eco_attempted, eco_won = extract_attempts_wins(first_team_cols[2])
                                semi_eco_attempted, semi_eco_won = extract_attempts_wins(first_team_cols[3])
                                semi_buy_attempted, semi_buy_won = extract_attempts_wins(first_team_cols[4])
                                full_buy_attempted, full_buy_won = extract_attempts_wins(first_team_cols[5])

                                all_econ_data.append([
                                    idx + 1, team_name, pistol_won,
                                    eco_attempted, eco_won,
                                    semi_eco_attempted, semi_eco_won,
                                    semi_buy_attempted, semi_buy_won,
                                    full_buy_attempted, full_buy_won
                                ])

                df_econ = pd.DataFrame(all_econ_data)

                # Merge Economy Data with Match Data by Row Order
                df_final = pd.concat([df.reset_index(drop=True), df_econ.reset_index(drop=True)], axis=1)

            except:
                df_final = df  # If no economy data, just save match results

            all_matches_data.append(df_final)

            # **Auto-Save after every match**
            pd.concat(all_matches_data, ignore_index=True).to_csv("all_vlr_matches.csv", index=False)

            print(f"✔ Match {i+1} on Page {current_page} Saved ✅")
            driver.back()
            time.sleep(3)

        except StaleElementReferenceException:
            continue
        except NoSuchElementException:
            continue

    try:
        next_button = driver.find_element(By.LINK_TEXT, str(current_page + 1))
        next_button.click()
        current_page += 1
        time.sleep(3)
    except:
        print("No more pages found. Exiting loop.")
        break

df_all_matches = pd.concat(all_matches_data, ignore_index=True)
df_all_matches.to_csv("all_vlr_matches.csv", index=False)
print("\n✅ All Matches Data Saved: all_vlr_matches.csv ✅")
```

```python

# Start Selenium
driver = gs.UndetectedChrome()
url = "https://vlr.gg/"
driver.get(url)

# Wait for the page to load
time.sleep(5)

# Click on the "Matches" tab
driver.find_element(By.LINK_TEXT, "Matches").click()
time.sleep(3)

# Click on the "Results" tab
driver.find_element(By.LINK_TEXT, "Results").click()
time.sleep(3)

# Get total number of pages
pagination = driver.find_elements(By.CSS_SELECTOR, "a[href*='?page=']")
total_pages = int(pagination[-1].text.strip()) if pagination else 1
print(f"Total Pages Found: {total_pages}")

# Initialize list to store all match data
all_matches_data = []

### **LOOP THROUGH MATCH PAGES**
current_page = 1
while current_page <= total_pages:
    print(f"\nScraping Page {current_page}/{total_pages}...")

    # Get updated match links each time before looping
    match_links = driver.find_elements(By.CSS_SELECTOR, "a.wf-module-item.match-item")

    # Loop through each match on the page
    for i in range(len(match_links)):
        try:
            match_links = driver.find_elements(By.CSS_SELECTOR, "a.wf-module-item.match-item")
            match = match_links[i]  # Get match at index i

            driver.execute_script("arguments[0].scrollIntoView();", match)  # Scroll to element if needed
            match.click()  # Click to open match page
            time.sleep(3)  # Wait for page to load

            # Extract match details
            match_soup = BeautifulSoup(driver.page_source, "html.parser")

            # Extract Team Names
            teams = match_soup.find_all("div", class_="team-name")
            team1_name = teams[0].text.strip() if len(teams) > 0 else "N/A"
            team2_name = teams[1].text.strip() if len(teams) > 1 else "N/A"

            # Extract Match Date
            date_element = match_soup.find("div", class_="match-header-date")
            match_date = date_element.text.strip().replace("\n", "").replace("\t", " ") if date_element else "N/A"
            match_date = " ".join(match_date.split())  # Remove duplicate spaces

            # Find all map sections
            map_sections = match_soup.find_all("div", class_="vm-stats-game")

            # Initialize list to store match data
            new_rows = []

            # Loop through each map section
            for map_section in map_sections:
                map_name = map_section.find("div", class_="map").text.strip() if map_section.find("div", class_="map") else "N/A"
                map_name = " ".join(map_name.split())  # Removes extra spaces, \n, \t

                # Extract scores and determine winner
                scores = map_section.find_all("div", class_="score")
                if len(scores) == 2:
                    team1_map_score = int(scores[0].text.strip())
                    team2_map_score = int(scores[1].text.strip())
                    team1_win = 1 if team1_map_score > team2_map_score else 0
                else:
                    team1_win = "N/A"

                # Append match data
                new_rows.append({"team1": team1_name, "team2": team2_name, "date": match_date, "map_name": map_name, "team1_win": team1_win})

            # Convert to DataFrame
            df = pd.DataFrame(new_rows)

            # Remove unplayed maps (where map_name or team1_win is "N/A")
            df = df[(df["map_name"] != "N/A") & (df["team1_win"] != "N/A")]

            ### **ECONOMY DATA EXTRACTION**
            try:
                # Click on the "Economy" tab
                econ_tab = driver.find_element(By.LINK_TEXT, "Economy")
                econ_tab.click()
                time.sleep(3)  # Allow the page to load

                # Get updated page source
                match_soup = BeautifulSoup(driver.page_source, "html.parser")

                # Locate all economy sections (one per map, limit to 3 max)
                active_map_sections = match_soup.find_all("div", class_="vm-stats-game")[:3]

                # Function to extract attempts & wins from "X (Y)" format
                def extract_attempts_wins(text):
                    text = text.strip().replace("\n", "").replace("\t", " ")  # Clean text
                    text = " ".join(text.split())  # Remove extra spaces
                    if "(" in text:
                        attempts, wins = text.split("(")[0].strip(), text.split("(")[1].replace(")", "").strip()
                    else:
                        attempts, wins = text, "0"  # If no parentheses, assume 0 wins
                    return int(attempts), int(wins)

                # Initialize list to store economy data
                all_econ_data = []

                # Loop through up to the first 3 map sections
                for idx, active_map_section in enumerate(active_map_sections):
                    econ_table = active_map_section.find("table", class_="wf-table-inset mod-econ")

                    if econ_table:
                        rows = econ_table.find_all("tr")

                        if len(rows) > 1:
                            first_team_row = rows[1]  # Index 1, since index 0 is the header
                            first_team_cols = [col.text.strip() for col in first_team_row.find_all("td")]

                            if len(first_team_cols) >= 6:
                                team_name = first_team_cols[0]
                                pistol_won = int(first_team_cols[1])

                                eco_attempted, eco_won = extract_attempts_wins(first_team_cols[2])
                                semi_eco_attempted, semi_eco_won = extract_attempts_wins(first_team_cols[3])
                                semi_buy_attempted, semi_buy_won = extract_attempts_wins(first_team_cols[4])
                                full_buy_attempted, full_buy_won = extract_attempts_wins(first_team_cols[5])

                                all_econ_data.append([
                                    idx + 1, team_name, pistol_won,
                                    eco_attempted, eco_won,
                                    semi_eco_attempted, semi_eco_won,
                                    semi_buy_attempted, semi_buy_won,
                                    full_buy_attempted, full_buy_won
                                ])

                df_econ = pd.DataFrame(all_econ_data)

                # Merge Economy Data with Match Data by Row Order
                df_final = pd.concat([df.reset_index(drop=True), df_econ.reset_index(drop=True)], axis=1)

            except:
                df_final = df  # If no economy data, just save match results

            all_matches_data.append(df_final)

            # **Auto-Save after every match**
            pd.concat(all_matches_data, ignore_index=True).to_csv("all_vlr_matches.csv", index=False)

            print(f"✔ Match {i+1} on Page {current_page} Saved ✅")
            driver.back()
            time.sleep(3)

        except StaleElementReferenceException:
            continue
        except NoSuchElementException:
            continue

    # Click Next Page
    try:
        next_button = driver.find_element(By.LINK_TEXT, str(current_page + 1))
        next_button.click()
        current_page += 1
        time.sleep(3)
    except:
        print("No more pages found. Exiting loop.")
        break

df_all_matches = pd.concat(all_matches_data, ignore_index=True)
df_all_matches.to_csv("all_vlr_matches.csv", index=False)
print("\n✅ All Matches Data Saved: all_vlr_matches.csv ✅")
```

```python
# Get updated page source
soup = BeautifulSoup(driver.page_source, "html.parser")

# Locate all economy sections (one per map, limit to 3 max)
active_map_sections = soup.find_all("div", class_="vm-stats-game")[:3]  # Limit to 3 maps

# Debugging: Print the economy table sections to check structure
print("\nExtracted Economy Sections Count:", len(active_map_sections))

# Function to extract attempts & wins from "X (Y)" format
def extract_attempts_wins(text):
    text = text.strip().replace("\n", "").replace("\t", " ")  # Clean text
    text = " ".join(text.split())  # Remove extra spaces
    if "(" in text:
        attempts, wins = text.split("(")[0].strip(), text.split("(")[1].replace(")", "").strip()
    else:
        attempts, wins = text, "0"  # If no parentheses, assume 0 wins
    return int(attempts), int(wins)

# Initialize list to store economy data
all_econ_data = []

# Loop through up to the first 3 map sections
for idx, active_map_section in enumerate(active_map_sections):
    print(f"\nProcessing Economy Data for Map {idx + 1}...")

    # Find economy table inside this map section
    econ_tables = active_map_section.find_all("table", class_="wf-table-inset mod-econ")

    if not econ_tables:
        print(f"Warning: No economy table found for Map {idx + 1}. Skipping...")
        continue  # Skip this iteration

    print(f"Found {len(econ_tables)} economy tables for Map {idx + 1}")

    # Use the first table if multiple are found
    econ_table = econ_tables[0]

    rows = econ_table.find_all("tr")

    if len(rows) <= 1:  # No valid data rows
        print(f"Warning: No valid rows found in economy table for Map {idx + 1}.")
        continue

    first_team_row = rows[1]  # Index 1, since index 0 is the header
    first_team_cols = [col.text.strip() for col in first_team_row.find_all("td")]

    if len(first_team_cols) < 6:  # Ensure the row has enough columns
        print(f"Warning: Economy data format unexpected for Map {idx + 1}. Skipping...")
        continue

    # Extract economy data
    team_name = first_team_cols[0]
    pistol_won = int(first_team_cols[1])
    eco_attempted, eco_won = extract_attempts_wins(first_team_cols[2])
    semi_eco_attempted, semi_eco_won = extract_attempts_wins(first_team_cols[3])
    semi_buy_attempted, semi_buy_won = extract_attempts_wins(first_team_cols[4])
    full_buy_attempted, full_buy_won = extract_attempts_wins(first_team_cols[5])

    # Append extracted data
    all_econ_data.append([
        idx + 1, team_name, pistol_won,
        eco_attempted, eco_won,
        semi_eco_attempted, semi_eco_won,
        semi_buy_attempted, semi_buy_won,
        full_buy_attempted, full_buy_won
    ])

# Define column names (removing "Pistol Attempted")
columns = [
    "Map Index", "Team", "Pistol Won",
    "Eco Attempted", "Eco Won",
    "Semi-Eco Attempted", "Semi-Eco Won",
    "Semi-Buy Attempted", "Semi-Buy Won",
    "Full-Buy Attempted", "Full-Buy Won"
]

# Convert to DataFrame
df_econ = pd.DataFrame(all_econ_data, columns=columns)

# **Check if df_econ is empty before saving**
if df_econ.empty:
    print("Warning: No economy data extracted. Only saving match results.")
else:
    df_econ.to_csv("economy_data_cleaned.csv", index=False)
    print("\nEconomy data saved: economy_data_cleaned.csv")

# Print confirmation
print(df_econ)
```

```python
# Click on the "Economy" tab
econ_tab = driver.find_element(By.LINK_TEXT, "Economy")
econ_tab.click()
time.sleep(3)  # Wait for the page to load

# Get updated page source
soup = BeautifulSoup(driver.page_source, "html.parser")

# Locate economy sections (one per map)
econ_sections = soup.find_all("table", class_="wf-table-inset mod-econ")

# Ensure economy stats align with maps
econ_data = []

for i, econ_table in enumerate(econ_sections):
    rows = econ_table.find_all("tr")[1:]  # Skip headers

    # Function to clean extracted economy values
    def clean_text(text):
        text = text.strip().replace("\n", "").replace("\t", " ")  # Remove newlines and tabs
        text = " ".join(text.split())  # Remove extra spaces
        text = text.split("(")[0].strip()  # Remove numbers inside parentheses
        return text

    # Extract and clean economy values for team1
    cols_team1 = rows[0].find_all("td")
    team1_pistols_won = clean_text(cols_team1[1].text)
    team1_eco_won = clean_text(cols_team1[2].text)
    team1_semi_eco = clean_text(cols_team1[3].text)
    team1_semi_buy = clean_text(cols_team1[4].text)
    team1_full_buy = clean_text(cols_team1[5].text)

    # Extract and clean economy values for team2
    cols_team2 = rows[1].find_all("td")
    team2_pistols_won = clean_text(cols_team2[1].text)
    team2_eco_won = clean_text(cols_team2[2].text)
    team2_semi_eco = clean_text(cols_team2[3].text)
    team2_semi_buy = clean_text(cols_team2[4].text)
    team2_full_buy = clean_text(cols_team2[5].text)

    # Append economy data for each map
    econ_data.append({
        "map_name": df.loc[i, "map_name"],  # Ensure the map matches the right economy data
        "team1_pistols_won": team1_pistols_won,
        "team2_pistols_won": team2_pistols_won,
        "team1_eco_won": team1_eco_won,
        "team2_eco_won": team2_eco_won,
        "team1_semi_eco": team1_semi_eco,
        "team2_semi_eco": team2_semi_eco,
        "team1_semi_buy": team1_semi_buy,
        "team2_semi_buy": team2_semi_buy,
        "team1_full_buy": team1_full_buy,
        "team2_full_buy": team2_full_buy
    })

# Convert to DataFrame
df_econ = pd.DataFrame(econ_data)

# Merge economy data with the main DataFrame (sorted by map)
df = df.merge(df_econ, on="map_name")

# Save updated DataFrame
df.to_csv("vlr_match_data.csv", index=False)

# Print confirmation
print(df)
```
